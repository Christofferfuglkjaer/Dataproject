Cleaning the Data

```python
import pandas as pd

# Read data
df = pd.read_csv('Original data.csv', sep=';', header=0, decimal=',')

# Remove unnecessary columns
df = df.drop(['Center','Kassenr', 'ScandcleftID', 'Unnamed: 3', 'Født', 'Materialedato A',
       'Total Row Score A', 'Goslon Score A', 'Materialedato B',
       'Total Row Score B','Goslon Score B', 'Materialedato C',
       'Anteroposterior 1.2', 'Anteroposterior 2.1', 'Vertical 1.2', 
       'Vertical 2.2', 'Transverse.2','Crowding .2', 'Spacing.2', 
       'Tooth shape/size\r\n.1', 'Pan 1.2','Pan 2\r\n.2', 
       'Goslon Score C', 'Pan 1', 'Pan 1.1'], axis=1)

# Remove rows with missing values
df = df.dropna()

df = df.rename(columns={'Anteroposterior 1.1': 'Anteroposterior 2.1'})
df = df.rename(columns={'Anteroposterior 2': 'Anteroposterior 1.2'})
df = df.rename(columns={'Anteroposterior 1': 'Anteroposterior 1.1'})
df = df.rename(columns={'Anteroposterior': 'Anteroposterior 2.2'})

df = df.rename(columns={'Vertical 2.1': 'Vertical 2.2'})
df = df.rename(columns={'Vertical 1.1': 'Vertical 2.1'})
df = df.rename(columns={'Vertical 1': 'Vertical 1.1'})
df = df.rename(columns={'Vertical 2': 'Vertical 1.2'})

df = df.rename(columns={'Transverse': 'Transverse 1'})
df = df.rename(columns={'Transverse.1': 'Transverse 2'})

df = df.rename(columns={'Crowding ': 'Crowding 1'})
df = df.rename(columns={'Crowding .1': 'Crowding 2'})

df = df.rename(columns={'Spacing': 'Spacing 1'})
df = df.rename(columns={'Spacing.1': 'Spacing 2'})

df = df.rename(columns={'Tooth shape/size  \r\n': 'Tooth shape/size 1'})
df = df.rename(columns={'Tooth shape/size\r\n': 'Tooth shape/size 2'})

df = df.rename(columns={'Pan 2\r\n': 'Pan 1'})
df = df.rename(columns={'Pan 2\r\n.1': 'Pan 2'})


# Save
df.to_csv('Behandlet data.csv', sep=';', decimal=',')
df
```
